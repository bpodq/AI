```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100  # number of samples every group
nGroup = 6   # number of 
nRow = 2
nCol = 3
D = 10  # 间隔
X = np.zeros([n*nGroup, 2])
Y = np.zeros(n*nGroup)
y = 1
for i in range(nRow):
    for j in range(nCol):
        xk = np.random.randn(n, 2) + np.array([D*j, D*i])  # 这里要注意，i j与横纵坐标的关系
        X[(i*nCol+j)*n:(i*nCol+j+1)*n, :] = xk
        y *= -1
        Y[(i*nCol+j)*n:(i*nCol+j+1)*n] = y

plt.scatter(X[:, 0], X[:, 1], c=Y)
```

```python
dict = {'x1': X[:, 0], 'x2': X[:, 1], 'label': Y}
df = pd.DataFrame(dict)
df.to_csv('adaboost_train.csv', index=None)
```
