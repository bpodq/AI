```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.array([100, 100, 50, 50])  # number of samples every group
nRow = 2
nCol = 2

D = 10  # 间隔
X = np.zeros([np.sum(n), 2])
Y = np.zeros(np.sum(n))
y = [1, 0, 0, 1]
start = 0
end = n[0]
k = 0
for i in range(nRow):
    for j in range(nCol):
        xk = np.random.randn(n[k], 2) + np.array([D*j, D*i])  # 这里要注意，i j与横纵坐标的关系
        X[start:end, :] = xk
        Y[start:end] = y[k]
        if k<nRow*nCol-1:
            start += n[k]
            end += n[k+1]
            k += 1

plt.scatter(X[:, 0], X[:, 1], c=Y)
```

```python
dict = {'x1': X[:, 0], 'x2': X[:, 1], 'label': Y}
df = pd.DataFrame(dict)
df.to_csv('adaboost_train5.csv', index=None)
```
