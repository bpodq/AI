```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotnine as pn
from plotnine import *
```

```python
n = 10
X1 = np.random.randn(10, 2)
X2 = np.random.randn(10, 2)
Y1 = np.zeros(10)
Y2 = np.ones(10)

X = np.vstack((X1, X2))
Y = np.hstack((Y1, Y2))
```

```python
print(X)
print(Y)
print(X[:, 0])
```

```python
dict = {'x1': list(X[:, 0]), 'x2': list(X[:, 1]), 'Y': list(Y), 'Y2': list(Y+2)}
print(dict)
data = pd.DataFrame(dict)
```

```python
data.plot.scatter('x1', 'x2')
```

```python
data.plot.scatter('x1', 'x2', c=Y)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
```

```python
pn.ggplot(data, pn.aes('x1', 'x2', color='Y')) + pn.geom_point()
```

```python
ggplot(data, aes('x1', 'x2', fill='Y', color='Y')) + geom_point(size=3)
```

```python
ggplot(data, aes('x1', 'x2', fill='Y', color='Y')) + geom_point(aes(shape='Y'), size=3)  # shape
```

```python
ggplot(data, aes('x1', 'x2', color='Y')) + geom_line() + geom_point(shape='o', size=5)
```
